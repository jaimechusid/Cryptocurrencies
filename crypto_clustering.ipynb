### Deliverable 1: Preprocessing the Data for PCA

```python
# Load the crypto_data.csv dataset.
df_crypto = pd.read_csv('crypto_data.csv')
df_crypto
```

```python
# Keep all the cryptocurrencies that are being traded.
df_crypto = df_crypto[df_crypto['IsTrading'] == True]
df_crypto
```

```python
# Keep all the cryptocurrencies that have a working algorithm.
df_crypto = df_crypto[df_crypto['Algorithm'].notna()]
df_crypto
```

```python
# Remove the "IsTrading" column. 
df_crypto = df_crypto.drop(columns=['IsTrading'])
df_crypto
```

```python
# Remove rows that have at least 1 null value.
df_crypto = df_crypto.dropna(axis='index', how='any')
df_crypto
```

```python
# Keep the rows where coins are mined.
df_crypto = df_crypto[df_crypto['TotalCoinsMined'].notna()]
df_crypto
```

```python
# Create a new DataFrame that holds only the cryptocurrencies names.
new_crypto_df = df_crypto.set_index('Unnamed: 0')
new_crypto_df = new_crypto_df.rename_axis(index=None)
new_crypto_df
```

```python
# Drop the 'CoinName' column since it's not going to be used on the clustering algorithm.
new_crypto_df = new_crypto_df.drop(columns=['CoinName'])
new_crypto_df
```

```python
# Use get_dummies() to create variables for text features.
X = pd.get_dummies(new_crypto_df, columns=['Algorithm','ProofType'])
X
```

```python
# Standardize the data with StandardScaler().
X_scaled = StandardScaler().fit_transform(X)
```
